# 青空文庫の最新ランキングから1位の内容を取得する

取得元: [青空文庫 アクセスランキング](https://www.aozora.gr.jp/access_ranking/)

## 目標

- ランキング表を取得する（pandasのデータフレームに入れる）
- 最新の項目（本）について、XHTMLとテキスト形式を取得する
    - このノートブックで主に行う

```python
from selenium import webdriver
browser = webdriver.Chrome()
browser.get('https://www.aozora.gr.jp/access_ranking/')
```

```python
tables = browser.find_elements_by_tag_name('table')

# 2番目のtableに最新のランキング項目が入っている
table_index = 1
table = tables[table_index]
```

```python
# ランキング表を抽出する
trs = table.find_elements_by_tag_name('tr')

rank_list = []
for tr in trs:    
    tds = tr.find_elements_by_tag_name('td')
    str_date = tds[0].text
    str_url_xhtml = tds[1].find_element_by_tag_name('a').get_attribute('href')
    str_url_txt = tds[2].find_element_by_tag_name('a').get_attribute('href')
    
    rank_list.append([str_date, str_url_xhtml, str_url_txt])
```

```python
rank_list
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(rank_list, columns=['年月', 'URL (XHTML版)', 'URL (テキスト版)'])
df['タイムスタンプ'] = pd.to_datetime(df['年月'], format='%Y年%m月')

# 新しい方を上にする
df_sorted = df.sort_values(by='タイムスタンプ', ascending=False) 
df_sorted
```

## テキスト形式（zipファイル）の内容を取得してダウンロードする

```python
# ランキング1位の詳細ページに飛ぶ
browser.get(url)
```

```python
# リンクに「zip」が含まれる要素を抽出する（テキスト版）
extension = 'zip'
elems_link = browser.find_elements_by_partial_link_text(extension)
url_zip_content = elems_link[0].get_attribute('href')
url_zip_content
```

```python
# 参考: https://www.lifewithpython.com/2015/10/python-how-to-download-extact-zip-file.html

import requests

# @see http://stackoverflow.com/questions/16694907/how-to-download-large-file-in-python-with-requests-py

def download_file(url):
    """URL を指定してカレントディレクトリにファイルをダウンロードする
    """
    filename = url.split('/')[-1]
    r = requests.get(url, stream=True)
    with open(filename, 'wb') as f:
        for chunk in r.iter_content(chunk_size=1024):
            if chunk:
                f.write(chunk)
                f.flush()
        return filename

    # ファイルが開けなかった場合は False を返す
    return False
```

```python
zip_file = download_file(url_zip_content)
zip_file
```
